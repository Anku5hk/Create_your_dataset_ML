```
import cv2
import numpy as np
import imutils
from PIL import Image
import matplotlib.pyplot as plt
```

```
img = cv2.imread('/content/drive/My Drive/photos/face_new.jpg')
img.shape
```

```
plt.imshow(img)
plt.show()
```

```
# blur image (larger the kernel size larger the blur)
blur_img = cv2.GaussianBlur(img, (9,9), 0)
Image.fromarray(blur_img)
```

```
# roi image[startY:endY, startX:endX]

roi_img = img[20:80, 10:180]
Image.fromarray(roi_img)
```

```
# resize image 

qimg = imutils.resize(img, width=200)
Image.fromarray(qimg)
```

```
# rotate image

rimg = imutils.rotate(img, 45)
Image.fromarray(rimg)
```

```
# rotate whole image

rimg = imutils.rotate_bound(img, 45)
Image.fromarray(rimg)
```

```
# draw rect img, point(x,y), point2(x,y), color, thicknes

imgr = cv2.rectangle(img, (80, 50), (130, 120),(0,0,255),3)
Image.fromarray(imgr)
```

```
# draw circle -1 to fill circle
# img, point1, radius, color, thickness

cir_img = cv2.circle(img, (150, 50), 10, (0,0,255), 3)
Image.fromarray(cir_img)
```

```
# draw line point1, point2, color, thickness

c = cv2.line(img, (20, 50), (40, 60), (0,0,255), 3)
Image.fromarray(c)
```

```
# text img,text,point,font,scale,color,thickness

im = cv2.putText(img.copy(), "fee", (70, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0,255,0), 2)
Image.fromarray(im)
```

```
# gray img

gimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
Image.fromarray(gimg)
```

```
# threshold image to set specific values to a scaler say from 200-255 to 0
# img,from_pixel, to_pixel, mode_to set value(this one sets to 0)

er = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY_INV)[1]
Image.fromarray(er)
```
